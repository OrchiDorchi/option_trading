```python
import numpy as np
import pandas as pd

import talib as ta

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style('darkgrid')
```

```python
spy = pd.read_csv('spy_daily_2009_2022.csv', index_col=0)
spy.index = pd.to_datetime(spy.index)
spy
```

```python
intervals = [1, 2, 3, 5, 14, 21, 50]
for t in intervals:
    spy[f'f_ret_{t}'] = spy.Close.pct_change(t)

spy['f_natr'] = ta.NATR(spy['High'], spy['Low'], spy['Close'])
spy['f_rsi'] = ta.RSI(spy['Close'])
upper, middle, lower = ta.BBANDS(spy['Close'])

spy['f_norm_upper'] = upper/spy['Close']
spy['f_norm_lower'] = lower/spy['Close']
spy['f_norm_middle'] = middle/spy['Close']

spy.dropna(inplace=True)
spy.drop(['Open', 'High', 'Low', 'Volume'],axis=1,inplace=True)
spy
```

```python
spy['return'] = spy['Close'].pct_change().shift(-1)
```

```python
spy
```

```python
spy['target'] = np.where(spy['return'] > 0, 1, 0)
spy = spy[1:-1]
```
